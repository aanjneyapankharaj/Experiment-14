```python
# DataSet
import pandas as pd
import numpy as np
data={
    'Product':['Laptop','Mobile','Tablet','Headphones','Smartwatch','Camera'],
    'Price':[55000,50000,40000,5000,8000,45000],
    'Units_Sold':[25,60,35,80,50,15],
    'Discount':[10,15,12,20,18,8]
}
df=pd.DataFrame(data)
print("Original Dataset")
df

```

```python
#1. Min-Max Normlization
df['Price_MinMax']=(df['Price']-df['Price'].min())/(df['Price'].max()-df['Price'].min())
print("\nMin-Max Normalized Price")
print(df[['Product','Price','Price_MinMax']])
```

```python
# Normalization for Units Sold
df['Units_Sold_MinMax']=(df['Units_Sold']-df['Units_Sold'].min())/(df['Units_Sold'].max()-df['Units_Sold'].min())
print("\nMin-Max Normalized Units Sold")
print(df[['Product','Units_Sold','Units_Sold_MinMax']])
```

```python
# Normalization for Discount
df['Discount_MinMax']=(df['Discount']-df['Discount'].min())/(df['Discount'].max()-df['Discount'].min())
print("\nMin-Max Normalized Discount")
print(df[['Product','Discount','Discount_MinMax']])
```

```python
#2 Z-score Normalization
df['Units_Zscore']=(df['Units_Sold']-df['Units_Sold'].mean())/df['Units_Sold'].std()
print("\nZ-score Normalized Units Sold")
print(df[['Product','Units_Sold','Units_Zscore']])
```

```python
#3. Decimal Scaling
df['Price_Decimal']=df['Price']/100000
print("\nDecimal Scaled Price")
print(df[['Product','Price','Price_Decimal']])
```

```python
# Decimal Scaling for Disount
df['Discount_Decimal']=df['Discount']/100
print("\nDecimal Scaled Discount")
print(df[['Product','Discount','Discount_Decimal']])
```

```python
# Normalize Multiple Coloumns
cols=['Price','Units_Sold','Discount']
df_norm=(df[cols]-df[cols].min())/(df[cols].max()-df[cols].min())
print("\nNormalized Dataset")
print(df_norm)
```

# Part-B Data Conversion

```python
import pandas as pd
import numpy as np
data={
    'Order_ID':[101,102,103,104,105,106],
    'Customer_Gender':['Male','Female','Female','Male','Male','Female'],
    'Payment_Method':['Credit Card','UPI','UPI','Credit Card','Cash','Cash'],
    'Product_Category':['Electronics','Clothing','Electronics','Home','Books','Electronics'],
    'City':['Mumbai','Delhi','Bangalore','Chennai','Kolkata','Pune'],
    'Order_Value':[4500,2300,3200,5200,1800,4100]
}
df=pd.DataFrame(data)
df
```

```python
# Method 1: Label Encoding
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Customer_Gender_Encoded'] = le.fit_transform(df['Customer_Gender'])
print(df)
```

```python
# Method 2: One-Hot Encoding
df_encoded = pd.get_dummies(df, columns=['Payment_Method', 'Product_Category', 'City'])
print(df_encoded)
```

```python
# Method 3: Encoding Multiple Coloumns
df_multi=pd.get_dummies(df,columns=['Product_Category','City'])
print(df_multi)
```

```python
# Method 4: Dummy Encoding (Drop First)
df_dummy=pd.get_dummies(df,columns=['Customer_Gender','Payment_Method','Product_Category','City'],drop_first=True)
print(df_dummy)
```
